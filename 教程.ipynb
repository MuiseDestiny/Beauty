### 概述：
1. 这是一个根据导出的`树洞1v1`石墨表格（excel格式），自动逐行解析用户数据生成精心制作的图片，并以附件形式发送到用户邮箱的技术文档。

2. 上述功能已在所写Python代码中实现，具体来讲就是通过本机浏览器渲染一个html页面，并经过`html2canvas.js` https://github.com/niklasvh/html2canvas 将所需div元素渲染到canvas，最后生成图片缓存到本地的过程。

3. 因为模板固定，文字内容不固定难以匹配，所以采用这种方式。实现过程实验了Python的一些渲染html的为图片的库，遗憾的是没有任何一个可以达到要求，所以只能通过本地浏览器渲染这种相对繁琐的方式。尽管繁琐，它的效果的确非常好。首先实现了一个模板html，然后留了几个空，这些空就用上述表格里的信息来填写；然后用本地浏览器打开生成的html文件，一旦打开浏览器就会将生成的图片下载到本地（浏览器默认下载文件夹处，每台计算机可能有差异，比如我的是D:/Download）；最后读取这个图片并发送邮件。由于表格里的内容会有颜色强调，在实现读取这个颜色的过程中也有一些小坎坷，不过都已经解决了。通过实际操作，发现模板里的文字用<font color='yellow'>黄色</font>很难看清楚，建议后续编辑文字时候减少使用这种颜色。

4. 当然上面的这些我们不必关心，我们只需要运行这个`xxx.py`文件就可以了，运行之前需要搭建`Python`+`Selenium`的环境。Selenium是一个自动化测试网页的库，它通过一个本地的`driver.exe`文件操纵浏览器，从而实现我们的需求。另外一些难以爬取的网页也经常通过Selenium来实现爬虫。

本文档首先介绍`Python`的`Anaconda`环境搭建，`Anaconda`是一个Python+各种常用库的打包，直接安装它很方便；然后介绍`Selenium`库的安装；最后介绍了编译器和代码运行操作。
